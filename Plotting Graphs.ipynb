```python
import numpy as np
import matplotlib.pyplot as plt
import math

def fun(x):
    return x*math.pow(math.e,-(x*x))


t1=np.arange(-5,5.001,0.1)
g=[]
for t in t1:
    g.append(fun(t))
    
k=plt.plot(t1,g,"r-o")
plt.setp(k,markersize=1)
#plt.grid()
plt.show()
```
