#  Grade Point Average

The academic office has compiled data about students' grades. The data is provided as text from standard input in three parts: information about courses, information about students and information about grades. Each part has a specific line format, described below..

###### Information about courses

Line format: Course Code~Course Name~Semester~Year~Instructor

###### Information about students

Line format: Roll Number~Full Name

###### Information about grades

Line format: Course Code~Semester~Year~Roll Number~Grade

The possible grades are A, AB, B, BC, C, CD, D with corresponding grade points 10, 9, 8, 7, 6, 5 and 4. The grade point average of a student is the sum of his/her grade points divided by the number of courses. For instance, if a student has taken two courses with grades A and C, the grade point average is 8 = (10+6)รท2. If a student has not completed any courses, the grade point average is defined to be 0.


Each section of the input starts with a line containing a single keyword. The first section begins with a line containing Courses. The second section begins with a line containing Students. The third section begins with a line containing Grades. The end of the input is marked by a line containing EndOfInput.

Write a Python program to read the data as described above and print out a line listing the grade point average for each student in the following format:

Roll Number~Full Name~Grade Point Average

Your output should be sorted by Roll Number. The grade point average should be rounded off to 2 digits after the decimal point. Use the built-in function round().

```python
def calculate():
    
    for l in rollgrade:
        sum,count=0,0
        for gr in rollgrade[l]:
            if gr=='A':
                sum +=10
            elif gr=='AB':
                sum +=9
            elif gr=='B':
                sum +=8
            elif gr=='BC':
                sum +=7
            elif gr=='C':
                sum +=6
            elif gr=='CD':
                sum +=5
            elif gr=='D': # Mind it next time :)
                sum +=4
            else:
                sum=0
            count=count+1
        if count!=0:
            avg=sum/count
        else:
            avg=0
        rollavg[l]=round(avg,2)


if __name__=='__main__':
  
    rollname,rollgrade,rollavg={},{},{}

    inp=input()
    if inp=='Courses':
        while True:
            inp=input()
            if inp=='Students':
                break
            
    if inp=='Students':
        while True:
            inp=input()
            if inp=='Grades':
                break
            else:
                roll,name=inp.split('~')
                rollname[roll]=name.rstrip()
                rollgrade[roll]=[]
              
    if inp=='Grades':   
        while True:
            inp=input()
            if inp=='EndOfInput':
                break
            else:
                Ccode,sem,year,roll,grade=inp.split('~')
                rollgrade[roll].append(grade.rstrip())
          
    calculate()  
    listroll=list(rollname.keys())
    listroll=sorted(listroll)
    for l in listroll:
         print(l+"~"+rollname[l]+"~"+str(rollavg[l]))
```

```python
rollname,rollgrade,rollavg={},{},{}
```

```python
df=pd.DataFrame(columns=["RollNo","Full Name","Course Code","Semester","Year","Grade"])
print(df)
```

```python
# function to calculate grade as per criteria
def calculate():
    
    for l in rollgrade:
        sum,count=0,0
        for gr in rollgrade[l]:
            if gr=='A':
                sum +=10
            elif gr=='AB':
                sum +=9
            elif gr=='B':
                sum +=8
            elif gr=='BC':
                sum +=7
            elif gr=='C':
                sum +=6
            elif gr=='CD':
                sum +=5
            elif gr=='D':
                sum +=4
            else:
                pass
            count=count+1
            
        if count!=0:
            avg=sum/count
        else:
            avg=0
        rollavg[l]=round(avg,2)
```

```python
fh = open("data8.txt",'r') # opening file handler as fh

contents = fh.readlines() # contents holds the file as a list

start=contents.index('Students\n')
end=contents.index('Grades\n')

#for l in contents:
    #print(l,end="")
#print(start,end)
#print(contents[start:end])


# Reading values of section 2 and making dictionary and dataframe out of it
for l in contents[start+1:end]:
    roll,name=l.split('~')
    rollname[roll]=name.rstrip()
    rollgrade[roll]=[]
    # creation of data frame from data of file
    df=df.append({"RollNo":roll,"Full Name":name.rstrip(),"Grade":[]},ignore_index=True)
#print(rollname)

# Reading data of section 3
start=end
end=contents.index('EndOfInput')
for l in contents[start+1:end]:
    Ccode,sem,year,roll,grade=l.split('~')
    rollgrade[roll].append(grade.rstrip())
#print(rollgrade)
```
